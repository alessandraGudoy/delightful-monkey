1. Find the big $O$ of $f(x) = 3x^2 + log(x)^7$ 


First, we can ignore the constant term in the big O estimate.

As $x$ increases, $\log(x)^7$ grows much more slowly than $x^2$,
so the big O estimate is simply the dominant term, or $x^2$.

This can also be seen in the graph in the previous question, sinnce both functions are monotonically increasing for positive $x$

In this case, we can choose $C=3$ and $k=1$ as our witnesses
for all $x > 1$, we have:

$|3x^2 + \log(x)^7| \le 3|x^2|$

Therefore,  $f(x)=3x^2 + \log(x)^7$ is $O(x^2)$.

##### <span style="color:blue">Answer</span>

````python
def intersect(a, b):
    """ return the intersection of two lists """
    A = set(a)
    B = set(b)
    result = [k for k in A if k in B]   
    return set(result)

#test it
intersect({3,6,7,54,99,45},{5,7,3,56,45,98,42,3})
````

````python
def union(a, b):
    """ return the union of two lists """
    A = set(a)
    B = set(b)
    result = [k for k in A]
    for k in B:
        result.append(k) 
    return set(result)

union({3,6,7,54,99,45},{5,7,3,56,45,98,42,3})
````

````python

import random

def bubble_sort(name,arr):
    """ return the sorted list using bubble sort """
    for i in range(len(arr)):
        randdom_number = random.randint(0,100)
        print("%s: %s" % (name,randdom_number))
        for j in range(0, len(arr)-i-1):
            if arr[j] > arr[j+1] :
                arr[j], arr[j+1] = arr[j+1], arr[j]
    return arr

print(bubble_sort("Moseli",[43,2,6,22,55,2,76,21]))
````

Implement code to calcutale k factorial using python loops and if statements. Make it so that for each factorial answer, you return the answer __minus 47__ if $k!$ is even and you return the answer __plus 56__ if $k!$ is odd

Use $k=0,1,6,8,9,39,102$ to test your code

````python

###### iterative version#########

def factorial(k):
    final = 1
    if k == 0:
        final = 1
    else:
        result = 1
        for i in range(1,k+1):
            result = result * i
        final= result
    if final%2 == 0:
        return final-47
    else:
        return final+56

print(factorial(1))
print(factorial(2))
print(factorial(4))
print(factorial(7))
print(factorial(8))
````
